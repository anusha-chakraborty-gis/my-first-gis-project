```python
import geopandas as gpd
from shapely.geometry import Polygon
import matplotlib.pyplot as plt

# West Bengal এর আসল বর্ডারের কাছাকাছি
wb_coords = [
    (85.80, 27.25), (86.70, 27.15), (87.90, 26.90), (88.10, 26.30), 
    (89.85, 26.70), (89.90, 24.50), (89.70, 22.00), (88.50, 21.50), 
    (87.00, 21.60), (85.80, 24.00), (85.80, 27.25)
]

wb_polygon = Polygon(wb_coords)
wb_gdf = gpd.GeoDataFrame({'State': ['West Bengal'], 'Created_By': ['Anusha']}, 
                          geometry=[wb_polygon], crs="EPSG:4326")

fig, ax = plt.subplots(1, 1, figsize=(8,10))
wb_gdf.plot(ax=ax, color='#FF6B35', edgecolor='#004E89', linewidth=3)
ax.set_title("Anusha Chakraborty's West Bengal Map\nFirst GIS Project using Python", 
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig('Anusha_WB_Map.png', dpi=300, bbox_inches='tight')
plt.show()

print("তুমি করে দেখালে Anusha! এটা তোমার প্রথম GIS Project 🔥")
```

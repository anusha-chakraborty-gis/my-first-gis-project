```python
import geopandas as gpd
import matplotlib.pyplot as plt
india_path = r"C:\Program Files\Taktak\GADM Download\gadm41_IND_1.shp"
india_gdf = gpd.read_file(india_path)
print("Total States + UT:", len(india_gdf))
print(india_gdf.columns)
india_gdf.head()
```

```python
india_gdf.plot(figsize=(10,12), cmap='jet', edgecolor='black')
plt.title("Anusha's India Map - All States & UTs", fontsize=20)
plt.axis('off')
plt.show()
```

```python
india_gdf.plot(figsize=(10,12), cmap='jet', edgecolor='black')
plt.title("Anusha's India Map - All States & UTs", fontsize=20)
plt.axis('off')
plt.savefig("Anusha_India_Rinbow_Map.png", dpi=300, bbox_inches='tight')
plt.show()
```
